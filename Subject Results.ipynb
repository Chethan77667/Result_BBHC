```python
import pandas as pd
import numpy as np

# Load the Excel file
file_path = "I Sem BCA Dec 2024-Jan 2025 updated one new.xlsx"
df = pd.read_excel(file_path, sheet_name='Sheet2', header=None)

subject_code = input("Enter the subject code (e.g., BCACACS101): ").strip()
results = []
usn = None
name = None

# Iterate through the rows
for i in range(len(df)):
    row = df.iloc[i]

    # Update USN if found
    if str(row[4]).strip().upper() == 'USN':
        usn = row[5]

    # Check if subject code is in the row
    if subject_code in row.values:
        col_idx = list(row).index(subject_code)

        # Try to extract marks from next row
        try:
            mark_entry = df.iloc[i + 1, col_idx]
            if isinstance(mark_entry, str) and '+' in mark_entry:
                parts = mark_entry.replace('-', '0').split('+')
                parts = [float(p.strip()) for p in parts if p.strip().replace('.', '', 1).isdigit()]
                total_marks = sum(parts)
            else:
                total_marks = np.nan
        except:
            total_marks = np.nan

    # Once CGPA is reached, backtrack for name and finalize record
    if str(row[4]).strip().upper() == 'CGPA':
        for j in range(i - 1, max(i - 10, 0), -1):
            candidate_row = df.iloc[j]
            for col in range(4, 2, -1):
                val = candidate_row[col]
                if isinstance(val, str) and val.strip():
                    val = val.strip()
                    if not val.upper().startswith("USN") and val.upper() not in ["SGPA", "RESULT", "CLASS", "TERM GRADE", "M.C.NO"]:
                        name = val
                        break
            if name:
                break

        if name and usn and not pd.isna(total_marks):
            results.append((usn, name.upper(), total_marks))

            # Reset for next student
            name = None
            usn = None
            total_marks = None

# Create a DataFrame and sort by Total Marks descending
df_results = pd.DataFrame(results, columns=["USN", "Name", "Total Marks"])
df_results = df_results.dropna(subset=["Total Marks"])
df_results = df_results.sort_values(by="Total Marks", ascending=False)

# Save to Excel
output_file = input("Enter the name of the Excel file to save (without .xlsx): ")
df_results.to_excel(f"{output_file}.xlsx", index=False, engine='openpyxl')

print(f"✅ Saved to {output_file}.xlsx (sorted by Total Marks)")
```

```python
import pandas as pd
from io import BytesIO
from IPython.display import display

# === USER INPUT ===
excel_path = "F:\\Result Analysis\\III Sem B.Com Dec Jan 2025 new.xlsx"  # Replace with actual file path
output_filename = "Semester_Result_Analysis"  # Output file name without extension

# === FUNCTIONS ===

def extract_subject_dict(file) -> dict:
    df_subjects = pd.read_excel(file, sheet_name='Sheet1', header=None)
    subjects_dict = {}
    for i in range(len(df_subjects)):
        row = df_subjects.iloc[i]
        subject_code = str(row[3]).strip()
        subject_name = str(row[4]).strip()
        if len(subject_code) == 10 and subject_code.isalnum():
            subjects_dict[subject_code.upper()] = subject_name
    return subjects_dict

def extract_usns(df):
    student_blocks = []
    for i in range(len(df)):
        if str(df.iloc[i, 4]).strip().upper() == 'USN':
            for col in range(5, len(df.columns)):
                val = df.iloc[i, col]
                if isinstance(val, str) and val.strip().upper().startswith('U'):
                    student_blocks.append((val.strip().upper(), i))  # (USN, start_row)
                    break
    return student_blocks

def get_result_for_usn_subject(block, input_usn, subject_code, subjects_dict):
    subject_code = subject_code.strip().upper()
    for i in range(len(block)):
        row = block.iloc[i]
        if subject_code in row.values:
            col_idx = list(row).index(subject_code)
            try:
                result_status = block.iloc[i + 10, col_idx]
            except:
                result_status = ""

            name = "Not Found"

            for i in range(len(block) - 1):
                if str(block.iloc[i, 4]).strip().upper() == 'USN':
                    val = str(block.iloc[i + 1, 4]).strip()
                    if val and not val.upper().startswith("U0") and len(val.split()) >= 1:
                        name = val
                    break

            return {
                "Name": name.upper(),
                "USN": input_usn,
                f"{subject_code} - {subjects_dict[subject_code]}": result_status
            }

    return {
        "Name": "Not Found",
        "USN": input_usn,
        f"{subject_code} - {subjects_dict[subject_code]}": "N/A"
    }

def get_gpa_result_total_percent(block, input_usn):
    sgpa = cgpa = result_status = total_marks = max_total = percentage = "Not Found"

    for i in range(len(block)):
        row = block.iloc[i]
        label = str(row[4]).strip().upper()
        label2 = str(row[6]).strip().upper()
        if label == 'SGPA':
            sgpa = row[5]
        elif label == 'CGPA':
            cgpa = row[5]
        elif label.startswith("RESULT:"):
            result_status = label.split(":")[-1].strip().upper()

    return {
        "USN": input_usn,
        "SGPA": sgpa,
        "CGPA": cgpa,
        "Result": result_status,
    }

def get_total_and_percentage(block):
    total_marks = max_total = percentage = None

    for i in range(len(block)):
        row = block.iloc[i]
        if "Total" in row.values:
            try:
                col_idx = list(row).index("Total")
                total_marks = block.iloc[i + 7, col_idx]
                max_total = block.iloc[i + 8, col_idx]

                if isinstance(total_marks, (int, float)) and isinstance(max_total, (int, float)) and max_total != 0:
                    percentage = round((total_marks / max_total) * 100, 2)
            except Exception as e:
                print(f"Error extracting totals: {e}")
            break

    return {
        "Total Marks": total_marks,
        "Max Marks": max_total,
        "Percentage": percentage
    }

def clean_to_int(val):
    if isinstance(val, float) and val.is_integer():
        return int(val)
    return val
# === MAIN PROCESSING ===

df = pd.read_excel(excel_path, sheet_name='Sheet2', header=None)
subjects_dict = extract_subject_dict(excel_path)
student_blocks = extract_usns(df)

print(student_blocks)
full_data = []

for idx, (usn, start_idx) in enumerate(student_blocks, start=1):
    block = df.iloc[start_idx-5:start_idx + 12]
    student_row = {"USN": usn, "Name": "Not Found"}

    for code in subjects_dict:
        result = get_result_for_usn_subject(block, usn, code, subjects_dict)
        student_row[f"{code} - {subjects_dict[code]}"] = result.get(f"{code} - {subjects_dict[code]}", "")
        if student_row["Name"] == "Not Found" and result["Name"] != "Not Found":
            student_row["Name"] = result["Name"]

    gpa_data = get_gpa_result_total_percent(block, usn)
    student_row["SGPA"] = gpa_data["SGPA"]
    student_row["CGPA"] = gpa_data["CGPA"]
    student_row["Result"] = gpa_data["Result"]

    # 🔁 Add Total & Percentage
    total_data = get_total_and_percentage(block)
    student_row["Total Marks"] = clean_to_int(total_data["Total Marks"])
    student_row["Max Marks"] = clean_to_int(total_data["Max Marks"])
    student_row["Percentage"] = total_data["Percentage"]

    full_data.append(student_row)


final_df = pd.DataFrame(full_data)
final_df.insert(0, "Sl. No", range(1, len(final_df) + 1))
subject_columns = [f"{code} - {subjects_dict[code]}" for code in subjects_dict]
final_columns = ["Sl. No", "Name", "USN"] + subject_columns + ["Result", "Total Marks", "Max Marks", "Percentage", "CGPA", "SGPA"]
final_df = final_df[final_columns]
# Convert and clean 'Total Marks' and 'Max Marks' — replace NaNs or invalid with 0 and make integers
final_df["Total Marks"] = pd.to_numeric(final_df["Total Marks"], errors="coerce").fillna(0).astype(int)
final_df["Max Marks"] = pd.to_numeric(final_df["Max Marks"], errors="coerce").fillna(0).astype(int)

# Optional: Round percentage to 2 decimals (only if it's numeric)
final_df["Percentage"] = final_df["Percentage"].apply(lambda x: round(x, 2) if isinstance(x, (float, int)) else 0.0)


# === DISPLAY & EXPORT ===
display(final_df)
```
